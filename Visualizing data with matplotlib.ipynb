# Visualizing data with matplotlib
Somtimes graphs provide the best way to visualize data

The matplotlib library allows you to draw graphs to help with visualization

If we want to visualize data, we will need to load some data into a DataFrame

```python
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
newfile=Path.joinpath(Path.cwd().parent,'CSV_files/lots_of_flight_data.csv')
delays_df=pd.read_csv(newfile)
delays_df.dropna(inplace=True)
delays_df.drop_duplicates(inplace=True)
X=delays_df.loc[:,['DISTANCE','CRS_ELAPSED_TIME']]
y=delays_df.loc[:,['ARR_DELAY']]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=20)
regressor=LinearRegression()
regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)
```

A common plot used in data science is the scatter plot for checking the relationship between two columns If you see dots scattered everywhere, there is no correlation between the two columns If you see somethign resembling a line, there is a correlation between the two columns

You can use the plot method of the DataFrame to draw the scatter plot

kind - the type of graph to draw
x - value to plot as x
y - value to plot as y
color - color to use for the graph points
alpha - opacity - useful to show density of points in a scatter plot
title - title of the graph

```python
import matplotlib.pyplot as plt
delays_df.plot(kind='scatter',x='DISTANCE',y='ARR_DELAY',color='black',alpha=0.3,title='correlation between distance and flight delay')
plt.show()
```

```python
plt.xlabel('Departure delay (in minutes)')
plt.ylabel('Arrival delay (in minutes)')
plt.title('Predict arrival delay')
delays_df.plot(kind='scatter',x='DEP_DELAY',y='ARR_DELAY',color='red',alpha=0.3,title='correlation between departure delay and arrival delay')
plt.show()
```

```python
plt.xlabel('Departure delay (in minutes)')
plt.ylabel('Arrival delay (in minutes)')
plt.title('Predict arrival delay')
plt.plot(X_test,y_pred,color='red',linewidth=2)
plt.scatter(x=delays_df['DEP_DELAY'],y=delays_df['ARR_DELAY'],color='blue',alpha=0.3)
plt.show()
```
